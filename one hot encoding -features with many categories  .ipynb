```python
import pandas as pd
```

```python
data=pd.read_csv("marcedesebenz.csv" ,usecols=['X1','X2','X3','X4','X5'])
data
```

```python
#counting the unique values in each column
for i in data.columns:
    print("unique values:data[",i,"]:",len(data[i].unique()))
#counting the unique values in the X1 column
print("unique values:data[x1]:",len(data.X1.unique())) 
```

```python
#getting the top 10 values 

#counting the categorial values in the X1 column
print(data.X1.value_counts())
print("=====================================================")
#sorting the values using the sort_values() in the decending order using the parameter ascending=False
print(data.X1.value_counts().sort_values(ascending=False))
print("=====================================================")
#using the list compresentation to store the values 
top_10_lables=[y for y in data.X1.value_counts().sort_values(ascending=False).head(10).index]
top_10_lables
```

```python
import numpy as np

#adding the column by checking 
#applying the one_hot_encoding on the  the X5 column on the aa value of the X5 column
def one_hot_encoding(df,feature,values):
    for i in values:
        df[feature+"_"+i]=np.where(df[feature]==i,1,0)
one_hot_encoding(data,"X5","aa")
data
```

```python
#appling to the one_hot_encoding to the top 10 repeated values  in the X1 column
one_hot_encoding(data,"X1",top_10_lables)
data
```

How can we do that in python?
