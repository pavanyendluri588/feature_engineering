Summary of Feature-engine’s imputers main characteristics

Transformer

Numerical variables

Categorical variables

Description

MeanMedianImputer()

√

×

Replaces missing values by the mean or median

ArbitraryNumberImputer()

√

x

Replaces missing values by an arbitrary value

EndTailImputer()

√

×

Replaces missing values by a value at the end of the distribution

CategoricalImputer()

√

√

Replaces missing values by the most frequent category or by an arbitrary value

RandomSampleImputer()

√

√

Replaces missing values by random value extractions from the variable

AddMissingIndicator()

√

√

Adds a binary variable to flag missing observations

DropMissingData()

√

√

Removes observations with missing data from the dataset

The CategoricalImputer() performs procedures suitable for categorical variables. From version 1.1.0 it also accepts numerical variables as input, for those cases were categorical variables by nature are coded as numeric.

```python
#importing the data
path="Housing_Prices_Dataset_test.csv"
data=pd.DataFrame(dict(x1=[np.nan,1,3,np.nan,5,np.nan,10,10,12,np.nan],
                           x2=['a',np.nan,'c',np.nan,np.nan,'c','q','s',np.nan,np.nan],
                           x3=[11,np.nan,23,np.nan,55,np.nan,np.nan,34,55,78]))
print(data)
#data=pd.read_csv(main_data)

print(data.columns)
for i in data.columns:
    print(data[i].isnull().value_counts().sort_values())
    print("\n\n")
data.shape
```

```python
#using the meanmedian inpytation
data1=data.copy()
#
#importing imputation the feature engineering module 
import feature_engine.imputation as imputation
impu=imputation.MeanMedianImputer()
impu.fit(data1)
data1=impu.transform(data1)
data1
```

```python
print(data1.columns)
for i in data1.columns:
    print(data[i].isnull().value_counts().sort_values())
    print(data1[i].isnull().value_counts().sort_values())
    print("\n\n")
```

```python
data1.shape
```

```python
#imputation using the mean n the dataset 
data2=data.copy()
impu1=imputation.MeanMedianImputer(imputation_method="median")
print(data2)
print("==========================================")
impu1.fit(data2)
print(data2)
print("=========================================")
data2=impu1.transform(data2)
print(data2)
```

```python
#ArbitraryNumberImputer 
#arbitatry number means
data3=data.copy()
impu2=imputation.ArbitraryNumberImputer()
impu2
```

```python
impu2.fit(data3)
data3
```

```python
data3=impu2.transform(data3)
data3
```

```python
#EndTailImputer
data4=data.copy()
impu3=imputation.EndTailImputer()
impu3
```

```python
print(data4)
impu3.fit(data4)
print("============================================")
print(data4)
```

```python
data4=impu3.transform(data4)
data4
```
