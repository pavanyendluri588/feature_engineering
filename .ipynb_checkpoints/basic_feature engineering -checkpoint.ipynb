# renaming columns

```python
#type 1 renaming using the rename  function 
#{old_column_name:new_colums_name}
#changing the single column name 
#if we dont use inplece=True means the changes will not save to dataset (implace=True) means change the dataset  and sae the changes 
df.rename(columns={'Cement (component 1)(kg in a m^3 mixture)':'Cement'},inplace=True)
#changing the multiple column names 

df.rename(columns={'Blast Furnace Slag (component 2)(kg in a m^3 mixture)':'Blast_Furnace_Slag',
       'Fly Ash (component 3)(kg in a m^3 mixture)':'Fly_Ash',
       'Water  (component 4)(kg in a m^3 mixture)':'Water',
       'Superplasticizer (component 5)(kg in a m^3 mixture)':'Superplasticizer',
       'Coarse Aggregate  (component 6)(kg in a m^3 mixture)':'Coarse_Aggregate',
       'Fine Aggregate (component 7)(kg in a m^3 mixture)':'Fine_Aggregate', 'Age (day)':'Age'},inplace=True)
```
