```python
#using the loc function 
#using the loc method
Import pandas as pd
data = {'name':['michael','louis','jack','jasmine'],
        'grades':[30,70,40,80],
        'result':['N/A','N/A','N/A','N/A']}

df = pd.DataFrame(data,columns=['name','grades','result'])

df.loc[df.grades>50,'result']='success'

df.loc[df.grades<50,'result']='fail'

print(df)
```

```python
df1=data.copy()
df1
df1['Order Priority_rank']='null'
#'Critical':5,'High':4, 'Medium':3,'Low':2, 'Not Specified':0
df1.loc[df['Order Priority']=='Critical','Order Priority_rank']=Order_Priority_values_dict['Critical']
df1.loc[df['Order Priority']=='High','Order Priority_rank']=Order_Priority_values_dict['High']
df1.loc[df['Order Priority']=='Medium','Order Priority_rank']=Order_Priority_values_dict['Medium']
df1.loc[df['Order Priority']=='Low','Order Priority_rank']=Order_Priority_values_dict['Low']
df1.loc[df['Order Priority']=='Not Specified','Order Priority_rank']=Order_Priority_values_dict['Not Specified']
```

```python
df1
```

```python
df2=data.copy()
df2['Order Priority_rank']=df2['Order Priority']
df2['Order Priority_rank']=df2['Order Priority_rank'].replace(Order_Priority_values_dict.keys(),Order_Priority_values_dict.values())
```
