```python
#making the index of the dataset based on the index_col as id column 
data_test=pd.read_csv('Housing_Prices_Dataset_test.csv',index_col='Id')
print("data_test.shape:",data_test.shape)
data_test_columns=[x for x in data_test.columns]
print("no of columns in data_test: ",len(data_test_columns))
for i in data_test_columns:
    print(i)
data_test.index
```
