```python
import pandas as pd 
import numpy as np 
```

```python
path="marcedesebenz.csv"
data= pd.read_csv(path)
data
```

```python
data_most_repeated_values=pd.DataFrame()
data_most_repeated_values
```

```python
#data=data[["X0","X1","X2","X3","X4","X5","X6","X8"]]
#data
```

```python
for i in data.columns:
    data_most_repeated_values[i]=[x for x in data[i].value_counts().sort_values(ascending=True).head(1).index]
data_most_repeated_values
```

```python
data_feame_nost_repeated=pd.DataFrame()
```

```python
for i in data_most_repeated_values.columns:
    data_feame_nost_repeated[i]=np.where(data[i]==data_most_repeated_values[i][0],1,0)
data_feame_nost_repeated
```
