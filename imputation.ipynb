# Complete Case Analusis(CCA):

CCA method is most populary known as "Listwise deletion"

CCA is a stright forward technique to handel the missing daata in the dataset.

CCA directly delete the remove the rows  that have missing data.

After applying CCA on the dataset .the dataset will not contain the even 1 single missing value



====================================================

we will apply CCA only:
    
when data is Missing At Random(MAR).

Missing data is completly removed from the table.

====================================================

Advantages:

Easy to implement 

No data manipulation is required on the dataset


====================================================


Limitations:

Deleted data can be informative.

Can lead to the deletion of a large part of the data.

Can create a bias in the dataset, if a large amount of a particular type of variable is deleted from it.

The production model will not know what to do with Missing data.

====================================================

When to Use:

Data is MAR(Missing At Random).

Good for Mixed, Numerical, and Categorical data.

Missing data is not more than 5% – 6% of the dataset.

Data doesn’t contain much information and will not bias the dataset.

# Arbitary Value imputation

Arbitary Value is the value not belongs (not linked) to the asset(column) or liability.

Arbitrary value is also called fictitious value.


this is most important imputation techinique 

Arbitary value imputation can handle the both numerical and categorial variable(variables,columns).

This technique states that we group the missing values in a column and assign them to a new value that is far away from the range of that column.
 
Mostly we use values like 99999999 or -9999999 or “Missing” or “Not defined” for numerical & categorical variables.


===============================================================

when we use Arbitary Value imputation:

data is not Missing At Ramdomly(NOT MAR)

Arbitary Value imputation is sutabile for all the types of the data.

===============================================================

Assumptions:

Data is not Missing At Random.

The missing data is imputed with an arbitrary value that is not part of the dataset or Mean/Median/Mode of data.


===============================================================
Advantages:

Easy to implement.

We can use it in production.

It retains the importance of “missing values” if it exists.




===============================================================

Disadvantages:

Can distort original variable distribution.

Arbitrary values can create outliers.

Extra caution required in selecting the Arbitrary value.


https://www.analyticsvidhya.com/blog/2021/06/defining-analysing-and-implementing-imputation-techniques/

```python

path='Loan_Eligibility_dataset/loan-test.csv'
data = pd.read_csv(path)
data1=data.copy()
#Series.fillna() function is used to fill NA/NaN values using the specified method.
for i in data1.columns:
    print(i,":",type(i))
    data1[i]=data1[i].fillna("Missing")

  
        
    
```

# mean Imputation

mean is apply to the numerical data.


In mean imputation the null or missing values are replaced by the mean value of the column.




https://thedatamonk.com/missing-value-treatment-mean-median-mode-knn-imputation-and-prediction/



https://dkstatisticalconsulting.com/mean-median-and-mode-when-to-use-each-and-why/

```python
path='Loan_Eligibility_dataset/loan-test.csv'
data = pd.read_csv(path)
data2=data.copy()
mean1=data2['ApplicantIncome'].mean()

data2['ApplicantIncome'].fillna(mean1)
```

# meadian imputation 


This is also used on the numerical data.


the data has the odd no of columns the the missing values are replaced by the middle value of the column.


the data has the even no of columns the missing values are  replaced by the both middle values mean  of the column.
